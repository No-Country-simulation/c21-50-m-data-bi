```python
# Se describe la composición de las variables categóricas del dataset. 

x = ["payment_type" # tipo de pago
     ,"customer_age" # edad del cliente
     ,"employment_status"# estado de empleabilidad
     ,"housing_status" # estado de vivienda actual
     ,"has_other_cards" # ¿tiene otras tarjetas?
     ,"source" # origen de la transacción
     ,"device_os" # tipo de sistema operativo
     ,"fraud_bool" # Aplicación fraudulenta
     ]

for element in x: 
    print(element)
    print(data_base[element].value_counts(normalize=True))
```
