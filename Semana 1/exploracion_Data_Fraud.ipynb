A partir de la visualización de la composición de las variables numéricas del dataset, se sugieren los siguientes posibles cambios o aspectos a tener en cuenta, al momento del entrenamiento del modelo de Machine Learning y la construcción de la visualización de los datos en el dashboard: 
 1. Se sugiere eliminar la columna "month" (mes), ya que no representa datos sobre el intento de fraude en sí, sino sólo el momento en el que se hizo la solicitud dentro del periodo de tiempo en el que se tomaron los datos. 
 2. Se sugiere cambiar los valores negativos [-Infinito,-1] de la/s columna/s "prev_address_months_count", "current_address_months_count", "intended_balcon_amount", "bank_months_count", "session_length_in_minutes", "device_distinct_emails_8w" a nulo o 0, ya que no representa un valor numérico, sino la ausencia del dato.
 3. Se sugiere revisar la especificidad de las siguientes columnas: "zip_count_4w", "velocity_6h", "velocity_24h", "velocity_4w", "bank_branch_count_8w", "date_of_birth_distinct_emails_4w" y su utilidad para el modelo de ML.
 4. En las columnas "credit_risk_score", "income" se podría agrupar los valores en categorías tales como "alto", "medio", "bajo", con el propósito de ser visualizado con mayor facilidad. 
 5. Si bien las columnas "email_is_free", "phone_home_valid", "phone_mobile_valid", "has_other_cards", "foreign_request", "keep_alive_session", "device_fraud_count" son numéricas, representan datos binarios, por lo que su interpretación debe estar ligada a dicha cualidad. 

```python
# Se revisan outliers para las variables numéricas. 

numericas = ["prev_address_months_count", "current_address_months_count", "intended_balcon_amount", "bank_months_count",
    "session_length_in_minutes", "device_distinct_emails_8w", "zip_count_4w", "velocity_6h", 
    "velocity_24h", "velocity_4w", "bank_branch_count_8w", "date_of_birth_distinct_emails_4w",
    "credit_risk_score", "income"
]


for element in numericas: 
    print(f"Proceso para {element}")
    # Se calcula el Q1 (Primer cuartil) y Q3 (Tercer cuartil).
    Q1 = data_base[element].quantile(0.25)
    Q3 = data_base[element].quantile(0.75)

    print(f"El primer cuartil para {element} es {Q1}")
    print(f"El tercer cuartil para {element} es {Q3}")

    # Se hace el cálculo del IQR (Rango Intercuartil).
    IQR = Q3 - Q1
    print(f"El rango intercuartil para {element} es {IQR}")

    # Se establecen los límites para encontrar a los outliers.
    inferior = Q1 - 1.5 * IQR
    superior = Q3 + 1.5 * IQR

    # Filtrado de outliers
    outliers = data_base[
        (data_base[element] < inferior) | (data_base[element] > superior)
    ]

    # Se observan los casos de outliers
    print(f"Los outliers identificados para {element} son:")
    print(outliers)
```
